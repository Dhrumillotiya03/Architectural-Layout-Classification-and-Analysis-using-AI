```python
import os
from PIL import Image
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader
import torchvision.models as models
import torchvision.transforms as transforms

device= "cpu"

# Define your CNN model (Feature Extractor)
class FeatureExtractor(nn.Module):
    def __init__(self):
        super(FeatureExtractor, self).__init__()
        # Define only the convolutional layers for feature extraction
        self.cnn = models.resnet18(pretrained=True)
        # Remove the fully connected layer (classification layer)
        self.cnn = nn.Sequential(*list(self.cnn.children())[:-1])

    def forward(self, x):
        # Forward pass through the convolutional layers
        return self.cnn(x)

# Initialize feature extractor model
feature_extractor = FeatureExtractor().to(device="cpu")

# Define custom dataset class
class CustomDataset(Dataset):
    def __init__(self, root_dir, transform=None):
        self.root_dir = root_dir
        self.images = sorted(os.listdir(root_dir))
        self.transform = transform

    def __len__(self):
        return len(self.images)

    def __getitem__(self, idx):
        img_name = self.images[idx]
        img_path = os.path.join(self.root_dir, img_name)
        img = Image.open(img_path).convert('RGB')  # Ensure images are in RGB format
        if self.transform:
            img = self.transform(img)
        # Extract label from the filename (e.g., 'class_label.jpg')
        label = int(img_name.split('.')[0])  # Extract class label from filename
        return img, label

# Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize images to a uniform size
    transforms.ToTensor(),  # Convert images to PyTorch tensors
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])  # Normalize with ImageNet statistics
])

# Define root directory of the dataset
root_dir = '/home/shravani_k/SEM6/DS-203/Project/E7-images'

# Create custom dataset
dataset = CustomDataset(root_dir=root_dir, transform=transform)

# Create DataLoader
batch_size = 32
dataloader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# Extract features from the dataset using the feature extractor
all_features = []
all_labels = []
for images, labels in dataloader:
    images = images.to(device)
    features = feature_extractor(images).squeeze()  # Remove the batch dimension
    all_features.append(features)
    all_labels.append(labels)
all_features = torch.cat(all_features, dim=0)
all_labels = torch.cat(all_labels, dim=0)

# Example usage of the extracted features and labels
print("Shape of extracted features:", all_features.shape)
print("Shape of labels:", all_labels.shape)
# Now, you can train a separate classifier using the extracted features and labels
```
